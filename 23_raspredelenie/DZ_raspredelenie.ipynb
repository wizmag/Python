# Задание 1

Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

%matplotlib inline
```

## Нормальное распределение

```python
plt.rcParams['figure.figsize'] = [10, 8]
for i,y in enumerate(['first', 'second', 'third', 'fourth', 'fifth'],0):
    mu = 2 + i
    sigma = 1 + i ** 0.5
    norm_rv = sts.norm(loc=mu, scale=sigma)
    norm_rv.rvs(size=10)
    x = np.linspace(0,10,100)
    cdf = norm_rv.cdf(x) 
    plt.plot(x, cdf, label=y)
    plt.ylabel('$F(x)$', fontsize = 20)
    plt.xlabel('$x$',fontsize = 20)

plt.legend(fontsize= 20)
plt.show()    
```

# Экспоненциальное распредление

```python
lam = 0.5
expon_rv = sts.expon(lam)
expon_rv.rvs(10)
x = np.linspace(0,5,100)
cdf = expon_rv.cdf(x)
ppf = expon_rv.ppf(x)
logcdf = expon_rv.logcdf(x)
logpmf = expon_rv.logpdf(x)
logsf = expon_rv.logsf(x)
plt.plot(x, cdf, label = 'cdf')
plt.plot(x, ppf,label = 'ppf')
plt.plot(x, logcdf,label = 'logcdf')
plt.plot(x, logpmf, label = 'logpmf')
plt.plot(x, logsf,label = 'logsf' )

plt.ylabel('$F(x)$',fontsize= 15)
plt.xlabel('$x$',fontsize= 15)
plt.legend(fontsize= 15)
plt.show()
```

# Распределение Стьюдента

```python
n = 10

t_rv = sts.t(n)

x = np.linspace(-5,5,100)
cdf = t_rv.cdf(x)
pdf = t_rv.pdf(x)
logcdf = t_rv.logcdf(x)
logpmf = t_rv.logpdf(x)
logsf = t_rv.logsf(x)
plt.plot(x, cdf, label = 'cdf')
plt.plot(x, pdf,label = 'pdf')
plt.plot(x, logcdf,label = 'logcdf')
plt.plot(x, logpmf, label = 'logpmf')
plt.plot(x, logsf,label = 'logsf' )

plt.ylabel('$F(x)$',fontsize= 15)
plt.xlabel('$x$',fontsize= 15)
plt.legend(fontsize= 15)
plt.show()
```
