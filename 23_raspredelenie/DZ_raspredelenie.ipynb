# Задание 1

Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

%matplotlib inline
```

## Нормальное распределение

```python
plt.rcParams['figure.figsize'] = [10, 8]
for i,y in enumerate(['first', 'second', 'third', 'fourth', 'fifth'],0):
    mu = 2 + i
    sigma = 1 + i ** 0.5
    norm_rv = sts.norm(loc=mu, scale=sigma)
    norm_rv.rvs(size=10)
    x = np.linspace(0,10,100)
    cdf = norm_rv.cdf(x) 
    plt.plot(x, cdf, label=y)
    plt.ylabel('$F(x)$', fontsize = 20)
    plt.xlabel('$x$',fontsize = 20)

plt.legend(fontsize= 20)
plt.show()    
```

# Экспоненциальное распредление

```python
for i,y in enumerate(['first', 'second', 'third', 'fourth', 'fifth'],2):
    mu = 2 + i
    sigma = 1 + i ** 0.5
    expon_rv = sts.expon(loc=mu, scale=sigma)    
    x = np.linspace(0,10,100)
    pdf = expon_rv.pdf(x)
    plt.plot(x, pdf, label = y)
    plt.ylabel('$F(x)$',fontsize= 15)
    plt.xlabel('$x$',fontsize= 15)
plt.legend(fontsize= 15)
plt.show()
```

# Распределение Стьюдента

```python
for i,y in enumerate(['first', 'second', 'third', 'fourth', 'fifth'],2):
    mu = 2 + i
    sigma = 1 + i ** 0.5
    x = np.linspace(-3,4,100)
    t_rv = sts.t(10,loc=mu, scale=sigma)
    pdf = t_rv.pdf(x)    
    plt.plot(x, pdf,label = y)
    

plt.ylabel('$F(x)$',fontsize= 15)
plt.xlabel('$x$',fontsize= 15)
plt.legend(fontsize= 15)
plt.show()
```
