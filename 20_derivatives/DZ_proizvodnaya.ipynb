```python
import matplotlib.pyplot as plt
import numpy as np
import pylab

from scipy.misc import derivative
from scipy.misc import electrocardiogram
from scipy.optimize import fmin, brute
from scipy.signal import find_peaks
```

# Уровень 0
посчитайте значение 1-ой и 2-ой производной фунции

f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1

```python
def f(x):
    return x**5 + 4 * np.sin(2*x) + np.cos(3*x+3)
```

```python
derivative(f,1,n=1)
```

```python
derivative(f,1,n=2)
```

```python
x1 = np.linspace(-5.0, 5.0, 100)
y1= [f(i) for i in x1]
d1= [derivative(f,i,n=1) for i in x1]
d2= [derivative(f,i,n=2) for i in x1]

plt.figure(figsize = (12,8))
plt.plot(x1, y1, 'r')
plt.plot(x1, d1, 'g')
plt.plot(x1, d2, 'b')
plt.title('График функции')
plt.ylabel('y')
plt.xlabel('x')

plt.show()
```

# Уровень 1**
Постройте график первой и второй производной функции

(sin(2x+1))^5
на отрезке [-5; 5]


```python
def func(x):
    return (np.sin(2*x + 1)) ** 5
```

```python
x1 = np.linspace(-5.0, 5.0, 100)
y1= [func(i) for i in x1]
d1= [derivative(func,i,n=1) for i in x1]
d2= [derivative(func,i,n=2) for i in x1]

plt.figure(figsize = (12,8))
plt.plot(x1, y1, 'r', label = 'Func')
plt.plot(x1, d1, 'g', label = 'Func_1')
plt.plot(x1, d2, 'b', label = 'Func_2')
plt.title('График функции')
plt.ylabel('y')
plt.xlabel('x')
plt.legend(loc = 'upper center')

plt.show()
```
