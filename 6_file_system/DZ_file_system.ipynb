2. Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки (если покупка была, сам файл visit_log.csv изменять не надо). Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории.

Учтите условия на данные:
- содержимое purchase_log.txt помещается в оперативную память компьютера
- содержимое visit_log.csv - нет; используйте только построчную обработку этого файла

````python
import json
````

````python
with open('purchase_log.txt', 'r', encoding = 'utf-8') as f2:
    purchase_dict = {}
    for line in f2:
        data = line.strip()
        log_dict = json.loads(data)
        purchase_dict[log_dict['user_id']] = log_dict['category']
purchase_dict
````

````python
with open('visit_log.csv', 'r', encoding = 'utf-8') as f:
    with open('funnel.csv', 'w', encoding = 'utf-8') as f2write:
        for line in f:
            data = f.readline().strip().split(',')
            if data[0] in purchase_dict:
                data.append(purchase_dict[data[0]])
                print(data)
                f2write.write(json.dumps(data)+'\n')
````
